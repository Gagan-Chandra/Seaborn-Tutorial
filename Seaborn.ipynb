```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
mpl.rcParams.update(mpl.rcParamsDefault)
%matplotlib inline
#Graph Styling
# https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html
plt.style.use('seaborn-darkgrid')
```

```python
def line_plot(df):
    n = int(input('Press 1 for Single line chart:\n' 
                  'Press 2 for Multi Line chart:\n'
                  'Press 3 for Comparision using hue\n'
                  ))
    if n == 1:
        print('select any column from', df.columns)
        sns.set(rc={"axes.facecolor":"#FFFDE7", "axes.grid":True,'xtick.labelsize':14,'ytick.labelsize':14})
        x = input("Enter X-axis column name:")
        y = input("Enter Y- axis column name:")
        c = input('Enter a color which is accepted by mathplotlib:')
        sns.lineplot(df[x],df[y],color = c)
        plt.xlabel(x) 
        plt.ylabel(y)
        plt.title('LINE PLOT')
        plt.show()
    elif n==2:
        print('select any column from', df.columns)
        x = input("Enter column name in x axis:")
        t = int(input("Enter number of columns in Y axis:"))
        y = []
        j = []
        colors=[]
        for i in range(t):
            d = input("Enter column name in y axis:")
            y.append(d)
            c = input('Enter a color which is accepted by mathplotlib:')
            colors.append(c)
        plt.figure(figsize=(12,5)) # Setting the figure size
        ax = plt.axes()
        ax.set_facecolor("darkgrey") # Setting the background color by using Hex code
        sns.set(rc={"axes.facecolor":"#FFFDE7", "axes.grid":True,'xtick.labelsize':14,'ytick.labelsize':14})
        for i in range(len(y)):
            sns.lineplot(x = df[x],y = df[y[i]],color = colors[i],label = y[i])
        plt.xlabel(x) 
        plt.ylabel(y)
        plt.legend(loc='upper right', fontsize='large',bbox_to_anchor=(1.2, 1.0))
        plt.title('LINE PLOT')
        plt.show()
    elif n==3:
        print('select any column from', df.columns)
        sns.set(rc={"axes.facecolor":"#FFFDE7", "axes.grid":True,'xtick.labelsize':14,'ytick.labelsize':14})
        x = input("Enter X-axis column name:")
        y = input("Enter Y- axis column name:")
        h = input('Enter hue:')
        sns.lineplot(x = x,y = y,hue = h, data = df)
        plt.xlabel(x) 
        plt.ylabel(y)
        plt.title('LINE PLOT')
        plt.show()
    elif n == 4:
        print('select any column from', df.columns)
        sns.set(rc={"axes.facecolor":"#FFFDE7", "axes.grid":True,'xtick.labelsize':14,'ytick.labelsize':14})
        x = input("Enter X-axis column name:")
        y = input("Enter Y- axis column name:")
        h = input('Enter hue:')
        f = input('Enter name of the column which u want to observe')
        sns.relplot(x = x,y = y,hue = h,col=f, data = df,kind='line',aspect=1)
        plt.xlabel(x) 
        plt.ylabel(y)
        plt.title('LINE PLOT')
        plt.show()
    else:
        print("Incorrect option")
        
        
data = input('Enter your csv file:')
df = pd.read_csv(data)
line_plot(df)
```

```python
def scatter_plot(df):
    n = int(input('Press 1 for one scatter plotting:\n'
                  'Press 2 for two scatter plottings:\n'
                  'Press 3 for morethan two scatter plottings:\n'
                  'Press 4 for hue:\n'
                  'Press 5 for grouping variables using hue:\n'))
    if n == 1:
        print('select any column from', df.columns)
        x = input("Enter X- axis column name:")
        y = input("Enter Y-axis column name:")
        c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
        sns.scatterplot(x = df[x],y = df[y] ,color = c)
        plt.xlabel(x , fontsize = 12)
        plt.ylabel(y , fontsize = 12)
        plt.title ('Scatter Graph')
        plt.show()
        
    if n == 2:
        print('select any column from', df.columns)
        x = input("Enter X- axis column name:")
        y = input("Enter Y-axis column name:")
        y1 = input("Enter another Y- axis column name:")
        c1 = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
        c2 = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
        sns.scatterplot(x = df[x],y = df[y] ,label = '1st scatter plot',marker='o' , color = c1)
        plt.xlabel(x , fontsize = 12)
        plt.ylabel(y , fontsize = 12)
        plt.title ('Scatter Graph')
        plt.show()
        sns.scatterplot(x = df[x],y = df[y1] ,label = '2nd scatter plot',marker='v',color =c2)
        plt.xlabel(x , fontsize = 12)
        plt.ylabel(y , fontsize = 12)
        plt.show()
        sns.scatterplot(x = df[x],y = df[y] ,label = '1st scatter plot',marker='o' , color = c1)
        sns.scatterplot(x = df[x],y = df[y1] ,label = '2nd scatter plot',marker='v',color =c2)
        plt.xlabel(x , fontsize = 12)
        plt.legend(loc='center right', bbox_to_anchor=(1.50, 0.5), ncol=1)
        z = y + y1
        plt.ylabel(z, fontsize = 12)
        plt.title ('Scatter Graph')
        plt.legend()
        plt.show()
    if n == 3:
        print('select any column from', df.columns)
        p = int(input("Enter number of plottings u want:"))
        y = []
        colors = []
        markers = ['o','v','s','p','*']
        x = input("Enter X- axis column name:")
        for i in range(p):
            b = input("Enter Y-axis column name:")
            y.append(b)
        colors = []
        for i in range(p):
            c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
            colors.append(c)
        for i in range(p):
            sns.scatterplot(x = df[x],y = df[y[i]],color =colors[i])
            plt.xlabel(x , fontsize = 12)
            plt.ylabel(y[i] , fontsize = 12)
            plt.title ('Scatter Graph')
            plt.show()
        for i in range(p):
            sns.scatterplot(x = df[x],y = df[y[i]] ,marker= markers[i],color =colors[i])
            plt.title ('Scatter Graph')
        plt.legend(loc='center right', bbox_to_anchor=(1.50, 0.5), ncol=1)
        plt.xlabel(x , fontsize = 12)
        plt.ylabel(y , fontsize = 12)
        plt.title ('Scatter Graph')
        plt.legend()
        plt.show()
    if n == 4:
        print('select any column from', df.columns)
        x = input("Enter X- axis column name:")
        y = input("Enter Y-axis column name:")
        h = input('Enter a hue column:')
        sns.scatterplot(x = x,y = y ,hue = h,data = df)
        plt.legend(loc='center right', bbox_to_anchor=(1.50, 0.5), ncol=1)
        plt.xlabel(x , fontsize = 12)
        plt.ylabel(y , fontsize = 12)
        plt.title ('Scatter Graph')
        plt.show()
    if n == 5:
        print('select any column from', df.columns)
        x = input("Enter X- axis column name:")
        y = input("Enter Y-axis column name:")
        h = input('Enter a hue column:')
        s = input('Enter a column name for grouping:')
        plt.figure(figsize=(10,10))
        plt.gcf().text(.5, .9, "Scatter Plot", fontsize = 40, color='Black' ,ha='center', va='center')
        sns.scatterplot(x = x,y = y ,hue = h,data = df)
        plt.legend(loc='center right', bbox_to_anchor=(1.50, 0.5), ncol=1)
        plt.xlabel(x , fontsize = 12)
        plt.ylabel(y , fontsize = 12)
        plt.title ('Scatter Graph')
        plt.show()
        
data = input('Enter your csv file:')
df = pd.read_csv(data)
scatter_plot(df)
```
